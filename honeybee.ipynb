```python
import cv2
import numpy as np
```

```python
image = cv2.imread('/content/sample_data/image4.jpg')
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_image)
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

```python
cv2_imshow(blurred_image)
```

```python
_, thresholded_image = cv2.threshold(blurred_image, 100, 255, cv2.THRESH_BINARY)
#_, thresholded_image = cv2.threshold(gray_image, 100, 255, cv2.THRESH_BINARY)
```

```python
cv2_imshow(thresholded_image)
```

```python
kernel = np.ones((5, 5), np.uint8)
morph_image = cv2.morphologyEx(thresholded_image, cv2.MORPH_CLOSE, kernel)
```

```python
cv2_imshow(morph_image)
```

```python
contours, _ = cv2.findContours(morph_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Draw contours on the original image for visualization
cv2.drawContours(morph_image, contours, -1, (0, 255, 0), 2)
```

```python
num_bees = len(contours)
```
